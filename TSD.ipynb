**TRAINING PROCESS AND SAVING THE MODEL**

```
import tensorflow as tf
import keras_preprocessing
from keras_preprocessing import image
from keras_preprocessing.image import ImageDataGenerator

TRAINING_DIR = "/content/Train"
training_datagen = ImageDataGenerator(
      rescale = 1./255,
	    rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')

VALIDATION_DIR = "/content/Train"
validation_datagen = ImageDataGenerator(rescale = 1./255)

train_generator = training_datagen.flow_from_directory(
	TRAINING_DIR,
	target_size=(40,40),
	class_mode='categorical',
  batch_size=126
)

validation_generator = validation_datagen.flow_from_directory(
	VALIDATION_DIR,
	target_size=(40,40),
	class_mode='categorical',
  batch_size=126
)

model = tf.keras.models.Sequential([
    # Note the input shape is the desired size of the image 150x150 with 3 bytes color
    # This is the first convolution
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(40, 40, 3)),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2),strides=(1,1),padding='valid'),
    # The second convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2),strides=(1,1),padding='valid'),
    # The third convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2),strides=(1,1),padding='valid'),

    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2),strides=(1,1),padding='valid'),
    # The fourth convolution
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2),strides=(1,1),padding='valid'),
    # Flatten the results to feed into a DNN
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),
    # 512 neuron hidden layer
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(15, activation='softmax')
])


model.summary()

model.compile(loss = 'categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

history = model.fit(train_generator, epochs=60, steps_per_epoch=35, validation_data = validation_generator, verbose = 1, validation_steps=3)

model.save("traffic.h5")
```

```
import numpy as np
from tensorflow import keras
from google.colab import files
from keras.preprocessing import image
from google.colab.patches import cv2_imshow
import cv2
model = keras.models.load_model('/content/drive/MyDrive/traffic.h5')
uploaded = files.upload()
danger_cascade = cv2.CascadeClassifier('/content/Traffic-sign-detection/xml_files/circular_lbp_new.xml')
for fn in uploaded.keys():
 
  # predicting images
  path = fn
  img = image.load_img(path, target_size=(40, 40))
  x = image.img_to_array(img)
  gray = cv2.cvtColor(x, cv2.COLOR_BGR2GRAY)
  gray = np.array(gray, dtype='uint8')
  danger, aaa, bbb=danger_cascade.detectMultiScale3(gray, scaleFactor=1.14, minNeighbors=18, outputRejectLevels = True)
  
  
  x = np.expand_dims(x, axis=0)
  
  images = np.vstack([x])
  classes = model.predict(images, batch_size=10)
  #print(fn)
  print(classes)


  if(classes[0][0]==1):
          text="100_KMPH"
  elif(classes[0][1]==1):
          text="120_KMPH"
  elif(classes[0][2]==1):
          text="20_KMPH"
  elif(classes[0][3]==1):
          text="30_KMPH"
  elif(classes[0][4]==1):
          text="50_KMPH"
  elif(classes[0][5]==1):
          text="60_KMPH"
  elif(classes[0][6]==1):
          text="80_KMPH"
  elif(classes[0][7]==1):
          text="CURVES"
  elif(classes[0][8]==1):
          text="KEEP_LEFT"
  elif(classes[0][9]==1):
          text="LEFT_TURN"
  elif(classes[0][10]==1):
          text="NARROW_ROAD"
  elif(classes[0][11]==1):
          text="RIGHT_TURN"
  elif(classes[0][12]==1):
          text="SCHOOLZONE"
  elif(classes[0][13]==1):
          text="SPEEDBREAKER"
  elif(classes[0][14]==1):
          text="STOP"
  else:
    text="not_trained"


  font = cv2.FONT_HERSHEY_SIMPLEX 
          
  org = (0, 24) 
        
  fontScale = 0.2
    
  color = (34, 139, 34) 
      
  thickness = 1
  print(text)
  image = cv2.putText(np.array(img), text, org, font, fontScale,color, thickness, cv2.LINE_AA, False)
  bigger = cv2.resize(image, (150, 150))
  for (x,y,w,h) in danger:
        cv.rectangle(bigger,(x,y),(x+w,y+h),(255,0,0),2)
  cv2_imshow(bigger)
  
```
